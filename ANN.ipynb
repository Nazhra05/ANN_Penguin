```python
import numpy as np
import pandas as pd
import sklearn
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv('/content/kenapa _kelompok_kami_yang_dapat_penguin.csv')

df
```

```python
print('Species', df['Species'].unique())
print('Island', df['Island'].unique())
print('Clutch', df['Clutch Completion'].unique())
print('Sex', df['Sex'].unique())
```

```python
new_df=df.copy()

new_df['Species']=new_df['Species'].replace({'Adelie Penguin (Pygoscelis adeliae)':'adeliae',
                                                 'Chinstrap penguin (Pygoscelis antarctica)':'antartica',
                                               'Gentoo penguin (Pygoscelis papua)':'papua'})

new_df['Sex'].fillna('.', inplace=True)
new_df['Sex']=new_df['Sex'].replace({'MALE':'male','FEMALE':'female'})
new_df['Sex']=new_df['Sex'].replace({'.':new_df['Sex'].mode()[0]})

new_df['Culmen Length (mm)'].fillna(new_df.groupby('Species')['Culmen Length (mm)'].transform('mean'),inplace=True)

new_df['Culmen Depth (mm)'].fillna(new_df.groupby('Species')['Culmen Depth (mm)'].transform('mean'),inplace=True)

new_df['Flipper Length (mm)'].fillna(new_df.groupby('Species')['Flipper Length (mm)'].transform('mean'),inplace=True)

new_df['Body Mass (g)'].fillna(new_df.groupby('Species')['Body Mass (g)'].transform('mean'),inplace=True)

new_df['Delta 15 N (o/oo)'].fillna(new_df.groupby('Species')['Delta 15 N (o/oo)'].transform('mean'),inplace=True)

new_df['Delta 13 C (o/oo)'].fillna(new_df.groupby('Species')['Delta 13 C (o/oo)'].transform('mean'),inplace=True)

new_df['Body Mass (kg)']=new_df['Body Mass (g)']/1000

new_df=new_df.drop(['Body Mass (g)','studyName','Sample Number','Comments','Individual ID','Region','Stage','Date Egg'], axis=1)


print(new_df.info())
new_df.head()
```

```python
print('Species', new_df['Species'].unique())
print('Island', new_df['Island'].unique())
print('Clutch', new_df['Clutch Completion'].unique())
print('Sex', new_df['Sex'].unique())
```

```python
from sklearn import preprocessing
LE=preprocessing.LabelEncoder()
```

```python
df_encode=new_df.copy()
df_encode['Island']=LE.fit_transform(df_encode['Island'])
df_encode['Clutch Completion']=LE.fit_transform(df_encode['Clutch Completion'])
df_encode['Sex']=LE.fit_transform(df_encode['Sex'])
df_encode.head()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
X=df_encode.drop(['Species'], axis=1)
Y=df_encode['Species']
bestfeatures = SelectKBest(score_func=f_classif, k='all')
fit = bestfeatures.fit(X,Y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)

featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Feature','Score']
print(featureScores.nlargest(9,'Score'))
```

```python
df_encode_drop=df_encode.copy()
df_encode_drop=df_encode_drop.drop(['Sex','Clutch Completion','Delta 13 C (o/oo)','Delta 15 N (o/oo)'], axis=1)
df_encode_drop.head()
```

```python
OHE=preprocessing.OneHotEncoder()
df_ohe=df_encode_drop.copy()

df_code=OHE.fit_transform(df_ohe[['Island']]).toarray()

df_list=list(sorted(new_df['Island'].unique()))

df_code=pd.DataFrame(df_code, columns=df_list)
df_ohe=pd.concat([df_code,df_ohe], axis=1)

df_ohe.head()
```

```python
df_ohe=df_ohe.drop(['Island'], axis=1)
df_ohe.head()
```

```python
df_ohe.shape
```

```python
df = df_ohe
```

```python
my_Species = df['Species'].value_counts()
Species_dict = {Species: idx for idx, Species in enumerate(list(my_Species.index))}
Species_dict
```

```python
df['Species'] = df['Species'].map(Species_dict)
df.head()
```

```python
df.shape
```

```python
df = df.sample(frac=1).reset_index(drop=True)
df.head()
```

```python
all_labels = df.pop('Species')
all_labels[: 9]
```

```python
df = (df - df.min())/(df.max() - df.min())
df.head()
```

```python
all_data = df.values
all_data[0]
```

```python
train_size = round(0.9 * all_data.shape[0])

train_data = all_data[: train_size]
test_data = all_data[train_size: ]

train_labels = all_labels[: train_size]
test_labels = all_labels[train_size: ]

assert(len(train_data) == len(train_labels))
```

```python
def build_model():
    model = keras.models.Sequential([
        keras.layers.Dense(64, 'selu'),
        keras.layers.Dense(64, 'selu'),
        keras.layers.Dropout(0.25),
        keras.layers.Dense(3, 'softmax'),
    ])

    model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['acc'])

    return model
```

```python
my_cb = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3,
        restore_best_weights=True)
```

```python
from sklearn.model_selection import StratifiedKFold

all_val_loss_histories = []
all_val_scores = []

skf = StratifiedKFold(n_splits=3)

for train_index, test_index in skf.split(train_data, train_labels):
    print('Fold...')
    x_train, x_val = train_data[train_index], train_data[test_index]
    y_train, y_val = train_labels[train_index], train_labels[test_index]

    model = build_model()

    history = model.fit(x_train, y_train, epochs=500,
                  validation_data=(x_val, y_val),
                  callbacks=[my_cb], verbose=1)

    all_val_loss_histories.append(history.history['val_loss'])

    all_val_scores.append(model.evaluate(x_val, y_val, verbose=0))
```

```python
avg_loss = np.mean([x[0] for x in all_val_scores])
print(f'Average loss is {avg_loss}')
```

```python
model = build_model()

history = model.fit(train_data, train_labels, epochs=26,
              callbacks=[my_cb], verbose=1)
```
